The code below was just to get a clear address of all the images. The images were split into 3 folders, we need to either merge the folders together or we get the address of all the files directing our path within each folder. I chose the latter, i.e creating an address for each of the images based on their subfolder.

Using a list comprehension method, i merged the folder name with the name of the file in that folder. The os.listdir simply returns the list of files in a folder. With this, i was able to create lists of addresses of each file in their respective folder

```python
first_image_set = ["part1/"+i for i in os.listdir("part1")]
second_image_set = ["part2/"+i for i in os.listdir("part2")]
third_image_set = ["part3/"+i for i in os.listdir("part3")]
```

Just checking some stats.

```python
print("The number of files in first set is",len(first_image_set))
print("The type of first set is",type(first_image_set))
print("The number of second in first set is",len(second_image_set))
print("The type of second set is", type(second_image_set))
print("The number of third in first set is", len(third_image_set))
print("The type of third set is",type(third_image_set))
```

Here i combined the lists into a single list. This super list now contains the paths of all the images in the dataset. In total we have 24,109 images.

```python
all_paths = first_image_set+second_image_set+third_image_set
print("There are",len(all_paths),"images in total")
```

If you truly checked out the dataset and visited the AICP GitHub repo, you'll find that I opened an issue in their repo regarding the dataset, I noticed some files with faulty name format.

Also, notice that the label of each image in the dataset is embedded in the name of the image. The naming format for each image is (age_gender_race_date). some of these files do not fulfill this naming strategy, so I had to find them and pull them out. Note how that the names have 3 underscores each and the first part which is the age is an integer. I'll take advantage of these facts to fish out faulty names.

```python
faulty=[] #empty list to store bad naming caught
for i in all_paths: # a for loop to go through all names
  try:                    # let's try this
    (i.split("_")[3])       # check if you can split name and get 3 underscores.
    int(i.split("_")[1])     # check if the age part is an integer
  except:                     # if you can do all the above fine, if you cant, do the following.
    faulty.append(i)                  #save the file name in the faulty list.
    
# Now that we are don with the loop, you can print the faulty lisy
print(faulty)
```

Remove faulty names from the main list of names.

```python
for i in faulty:
  all_paths.remove(i)
print("There are now",len(all_paths),"images in total")
```

Personally, I prefer my data in a Dataframe, so I'll set up the list as a dataframe using the pandas library.

```python
df = pd.DataFrame(all_paths, columns=["file_paths"]) # create dataframe and name the column "file_paths"
df.tail() #let's check out the tail.
```

In the next cell, what I did here was to split the labels into their various category, you'll find out eventually, that I used this same dataset for some other classification tasks, such as gender and race, but let's leave those discussions out of this module.

in this cell, I use string slicing to do this, by splitting the string, I'm able to cut out the part of the age, gender and race. Mind you they are all separated by an underscore as I've explained earlier, I took advantage of this to do the split.

```python
age = [int(i.split("_")[0][6:]) for i in all_paths]
gender = [int(i.split("_")[1]) for i in all_paths]
race = [int(i.split("_")[2]) for i in all_paths]
```

Now that we've gotten out the age, gender and race of each image, i simply added them to the dataframe.

```python
df["age"] = age
df["gender"] = gender
df["race"] = race
df.head()
```

Check out some info about the Dataframe.

```python
df.info()
```

In the Dataframe, and for a while now if you've been watching closely, you'll discover that the gender and race are in an integer form, the dataset source AICP already gave the description, that gender is 0 for male and 1 for female, they also gave the descriptions for the race too, do well to check it out yourself. Nevertheless, I used the format to map my gender and race columns in the DataFrame.

```python
df.gender = df.gender.map({0:"male",1:"female"}) # Rename the gender to words
df.race = df.race.map({0:"White",1:"Black",2:"Asian",3:"Indian",4:"Hispanic,Latino or Middle Eastern"}) #Rename to words
```

Just for my personal use, i decided to store the DataFrame as a csv file.

```python
df.head()
df.to_csv("data.csv", index=False)
```

###### Creating the Fastai DataBlock.

Remember to checkout the official [Fastai website](https://www.fast.ai/) to learn more about FastAI, as well as the docs.

Creating A datablock is quite easy, the datablock API is the foundational block of the fastai dataset creation, you'll have to tell it you are creating a dataset of images, and you'll be using it for a regression task.

The splitter simply splits your train set into train and validation, Fastai is very serious about overfitting, even if you don't set the splitter, it does that automatically for you.

Next, you'll have to tell the API how to fetch the address of the images, and the respective label for that image.Remember that I've all this in a nice dataframe, that's what we've spent the time doing so far

Finally, you can do some transformations for the images. Remember that to train a Convolutional neural network, images are best fitted in when they are of the same size. to this effect, I've told Fastai to crop all images to 224 by 224.

When you are done, create an instance of the datablock, the dataloader.

```python
human = DataBlock(
    blocks=(ImageBlock, RegressionBlock), 
    splitter=RandomSplitter(valid_pct=0.2, seed=2021),
  get_x = ColReader(0),
    get_y=ColReader(1),# 1 for age, 2 for gender, 3 for race.
item_tfms=Resize(224))
dls = human.dataloaders(df)
```

You can tell your dataloader to show you some of the validation set

```python
dls.valid.show_batch(max_n=5, nrows=1)
```

You can tell your dataloader to show you some of the Train set too.

```python
dls.train.show_batch(max_n=10, nrows=2)
```

Whooow! It's time to train the network. You'll find below, that I used the mobilenet_v2(by pytorch) architecture to train, you could try out other bigger network architectures from either fastai or pytorch, but it's always best to use the one moderate enough for your task so as not to overfit the data.

I wanted a very small architecture, that's why I used the mobilenet. The mobilenet is a Pytorch pretrained model, that's why i had to import the model from torchvision. You can also check out FastAI's resnet architectures, by swarping out the "models.mobilenet_v2" and putting resnet18, resnet50, or any other architecture.

since the model is pretrained on imagenet, this helps us to maximize the potential of the Network since it has been trained to identify different things before. What we just need to do is to fine-tune the model to our task.
The cnn_learner automatically passes the parameter "pretrained = True". If you don't want to use a pretrained model, you'll have to intentionally set it to false.

I'll advice you check out how fastai's learn.fine_tune works, but u'll see that i cut off the last layer of the pretrained model, so that i can train it to suit my own task.

Since I am training a regression model, i used the rsme score to watch my model performance, see as the model Root Mean Squared Error reduces as number of epochs increase.
For this task, I trained the model for 7 epochs.

```python
import torchvision.models as models
learn = cnn_learner(dls, models.mobilenet_v2, cut=-1,metrics=rmse)
learn.fine_tune(10)
```
